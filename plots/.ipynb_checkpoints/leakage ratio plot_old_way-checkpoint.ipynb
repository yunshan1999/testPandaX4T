```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
#plt.style.use('seaborn-whitegrid')
import numpy as np
from IPython.core.pylabtools import figsize # import figsize
#figsize(12.5, 4) # 设置 figsize
plt.rcParams['savefig.dpi'] = 200 #图片像素
plt.rcParams['figure.dpi'] = 200 #分辨率
```

```python
ELife = ([300, 400, 500, 600, 700, 800, 1100])
EEE = ([30, 40, 50, 60, 70, 80, 90])
SEG_res = ([0.2, 0.4, 0.6, 0.8])

LR_ELife = ([2.21193,0.942134,0.476674,0.285511,0.203938,0.162027,0.104824])
LR_EEE = [1.89162,1.07347,0.672127,0.445701,0.320738,0.241401,0.181128]
LR_SEG_res = ([0.247105,0.259054, 0.303614,0.415707,0.635353])

LR_err_ELife = ([0.0153614,0.00999896,0.00711138,0.00550383,0.00465181,0.00414637,0.00333489])
LR_err_EEE = ([0.0141727,0.0106724,0.00844458,0.0068765,0.00583352,0.00506113,0.00438398])
LR_err_SEG_res = ([0.00649603,0.00665115,0.00720039,0.00842612,0.010417])
```

```python
#plt.yscale("log")
plt.grid(True)
plt.xlabel("Electron life time (microsec)")
plt.ylabel("Leakage Ratio (%)")
x_coordinates = [600, 600]
y_coordinates = [0., 3.]
plt.plot(x_coordinates, y_coordinates,linestyle = "dotted",color = "orange")
plt.errorbar(ELife, LR_ELife, yerr=LR_err_ELife,color='black',linewidth = 1.,fmt = '.', capsize = 2)
plt.ylim(0.01,3.)
plt.legend()
plt.show()
```

```python
#plt.yscale("log")
plt.grid(True)
plt.xlabel(" Electron Extraction Efficiency(%)")
plt.ylabel("Leakage Ratio (%)")
x_coordinates = [72.7, 72.7]
y_coordinates = [0., 2.4]
plt.plot(x_coordinates, y_coordinates,linestyle = "dotted",color = "orange")
plt.errorbar(EEE, LR_EEE, yerr=LR_err_EEE,color='black',linewidth = 1.,fmt = '.', capsize = 2)
plt.ylim(0.,2.4)
plt.legend()
plt.show()
```

```python
#plt.yscale("log")
plt.grid(True)
plt.xlabel("SEG resolution")
plt.ylabel("Leakage Ratio(%)")
x_coordinates = [0.25, 0.25]
y_coordinates = [0.2, .7]
plt.plot(x_coordinates, y_coordinates,linestyle = "dotted",color = "orange")
plt.errorbar(SEG_res, LR_SEG_res, yerr=LR_err_SEG_res,color='black',linewidth = 1.,fmt = '.', capsize = 2)
plt.ylim(0.2,.7)
plt.legend()
plt.show()
```
